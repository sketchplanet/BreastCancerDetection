## Exploring Tumor patches

```python
import pandas as pd
from glob import glob
import os
import cv2
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline

image_path = "./image/"
```

### Plot
* ./image/normal 89,535
* ./image/tumor  31,355

```python
def subplots(prefix,filelist):
    plt.figure(figsize=(16, 12))
    ncol = 5
    nrow = math.ceil(len(filelist) // ncol)
    
    for i in range(0, len(filelist)):
        plt.subplot(nrow, ncol, i + 1)
        img = cv2.imread(os.path.join(prefix,filelist[i]))
        plt.imshow(img)
```

```python
tumor_filelist = os.listdir(os.path.join(image_path,'tumor'))
normal_filelist = os.listdir(os.path.join(image_path,'normal'))
```

## Tumor Patches

```python
#help(np.random.choice)
#print(np.random.choice(tumor_filelist,20))
subplots(os.path.join(image_path,'tumor'),np.random.choice(tumor_filelist,20))
```

## Normal Patches

```python
subplots(os.path.join(image_path,'normal'),np.random.choice(normal_filelist,20))
```
